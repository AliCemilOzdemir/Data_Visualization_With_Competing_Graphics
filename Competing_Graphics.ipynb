## Importing Essential Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import bar_chart_race as bcr
import numpy as np
```

### Uploading Data

```python
df = pd.read_csv("corona_dat.csv") # daily number of coronavirus cases of some countries
```

### Exploratory Data Analysis

we are going to look at these countries data :

**['Turkey','Italy','Spain','Germany','China','Russia','US']**

so lets take a closer look at this data: 

```python
df2=df[["date",'Turkey','Italy','Spain','Germany','China','Russia','US']]
df2.set_index('date',inplace=True)

sum_daily=df2.cumsum(axis=0)

df2.plot(figsize=(9,5), xlabel='Date', ylabel='Daily Value of CoronaVirüs Cases')
sum_daily.plot(figsize=(9,5), xlabel='Date', ylabel='Total Value of CoronaVirüs Cases');

plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

As it can be seen from the graphic,there is a radical increase at US's total number of coronavirus cases !

# Towards competing graphics
###### Now we have an idea of what kind of data we have ,and it's time to take an action to create a competing graphics for data visualization...

```python
# firstly ,let's see what is the type of df["date"] Series
df["date"].dtype
```

```python
df['date']=pd.to_datetime(df['date'])
df['date'].dtype # the output means that now we have a datetime64 datatype,dtype('<M8[ns]') = datetime64.
```

```python
df_competing_graph = df[["date",'Turkey','Italy','Spain','Germany','China','Russia','US']]
df_competing_graph.set_index('date',inplace=True)

df_competing_graph
```

<div class="alert alert-block alert-success">
<b>We've done a great progress !</b>


We want to see the data from start to end cumulatively ,but now what we have is daily reports of each day...
So let's make data cumulatively for each day...
</div>

```python
df_competing_graph=df_competing_graph.cumsum(axis=0)
df_competing_graph.head()
```

```python
# now if you have bar_chart_race and ffmpeg on your computer ,You can create a competing graphics of total value of coronavirus...
# it will be created at your current directory,if you used relative path to "filename" parameter.
# but if you used absolute path ,it will be created there...
bcr.bar_chart_race(df_competing_graph,filename='covid-19alicemilozdemir.mp4', 
                   figsize=(10,8), title='COVID- Çalışması @Ali Cemil Özdemir ')
```
